```python
t1 = torch.load(dir1+'/model_0_2664486226/best.pth.tar')
t2 = torch.load(dir2+'/model_0_2664486226/best.pth.tar')
t1.keys()
```

```python
#Validate that trained models are the same
def check_fields(d1,d2,key,tol=1e-5):
    print(key)
    # Check if values in key match between dictionaries
    if isinstance(d1[key],dict):
        for k in d1[key].keys():
            if key=='optim_dict':
                print(k,d1[key][k])
                print(k,d2[key][k])
            else:
                print(k,np.allclose(d1[key][k],d2[key][k],rtol=tol))
    else:
        print(d1[key],d2[key],np.allclose(d1[key],d2[key],rtol=tol))
    print('')
check_fields(t1,t2,'epoch')
check_fields(t1,t2,'stats')
check_fields(t1,t2,'best_val_loss')
check_fields(t1,t2,'state_dict')
check_fields(t1,t2,'optim_dict')
```
